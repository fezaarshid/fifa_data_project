# Which Country has the most number of players

```python
df=pd.read_csv("fifa_data.csv")
pd.set_option("display.max_columns",None)
df
```

```python
df["Nationality"].value_counts().idxmax()
```

# find the tallest player using the height column?

```python
df['Height']
```

```python
df['Height_inches'] = (
    df['Height']
    .dropna()
    .str.split("'", expand=True)
    .astype(int)
    .dot([12, 1])
)
```

```python
a = df.loc[
    df['Height_inches'] == df['Height_inches'].max(),
    ['Name', 'Height']
]
a
```

```python
import matplotlib.pyplot as plt
```

```python

foot_counts = df["Preferred Foot"].value_counts()

plt.figure()
plt.bar(foot_counts.index, foot_counts.values)
plt.xlabel("Preferred Foot")
plt.ylabel("Number of Players")
plt.title("Preferred Foot of FIFA Players")
plt.show()
```

# plot a scatter plot on overall and potential

```python
import matplotlib.pyplot as plt

plt.scatter(df['Overall'], df['Potential'])
plt.xlabel('Overall')
plt.ylabel('Potential')
plt.show()
```

# plot a histogram for salary range of players
